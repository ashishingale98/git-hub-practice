#### importing libraries

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.graph_objects as go
import pandas as pd
from matplotlib.pyplot import rcParams
rcParams['figure.figsize'] = 15,6
#from fbprophet import Prophet
#import pycountry
import plotly.express as px
from collections import namedtuple
```

#### importing data

```python
df=pd.read_csv("COVID_CLEAN.csv",parse_dates=['Date'], index_col='Date')
df.rename(columns={"Country/Region":"Country"},inplace=True)
```

```python
df.head()
```

```python
df=df.groupby("Date")[['Confirmed', 'Deaths', 'Recovered']].sum()
```

```python
df
```

#### divind into confirmed, deaths adn recoveered cases

```python
c_df=df
```

```python
c_df=c_df.drop(['Deaths','Recovered'],axis=1)
c_df.head()
```

```python
c_df['Confirmed'].plot(kind="line",figsize=(15,5))
plt.show()
```

```python
c_df.shape
```

```python
c_df['Daily']=c_df['Confirmed'].diff().dropna()
```

```python
c_df=c_df.dropna()
c_df.head()
```

```python
c_df.shape
```

```python
plt.plot(c_df['Daily'])
plt.xticks(rotation=90)
plt.show()
```

```python
# To to a 70 - 30 split 
size = int(len(c_df['Daily'])*0.7)  # sets the size or index of 80% of the data
size

### Splitting the Data
df_train = c_df.iloc[:size]
df_test = c_df.iloc[size:]
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
#Perform Dickey-Fuller test:
from statsmodels.tsa.stattools import adfuller

print ('Results of Dickey-Fuller Test:')
dftest = adfuller(df_train['Daily'], autolag='AIC') 

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
print(dfoutput)
```

### Seasonality

```python
# Decomposing the series
import statsmodels.tsa.seasonal as sts
decomposition = sts.seasonal_decompose(df_train['Daily'],model='multiplicative', freq=3)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
df_train['Daily'].plot(kind="line",figsize=(10,6),label='Original')
plt.subplot(412)
trend.plot(kind="line",figsize=(10,6),label='trend')
plt.legend(loc='best')
plt.subplot(413)
seasonal.plot(kind="line",figsize=(10,6),label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
residual.plot(kind="line",figsize=(10,6),label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
diff1=c_df['Daily'].diff()
diff1=diff1.dropna()
```

```python
#Perform Dickey-Fuller test:
from statsmodels.tsa.stattools import adfuller

print ('Results of Dickey-Fuller Test:')
dftest = adfuller(diff1, autolag='AIC') 

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
print(dfoutput)
```

```python
diff2=diff1.diff()
diff2=diff2.dropna()
```

```python
#Perform Dickey-Fuller test:
from statsmodels.tsa.stattools import adfuller

print ('Results of Dickey-Fuller Test:')
dftest = adfuller(diff2, autolag='AIC') 

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
print(dfoutput)
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
```

```python
from pandas import Series
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from matplotlib import pyplot
pyplot.figure(figsize=(15,10))
pyplot.subplot(211)
plot_acf(diff2, ax=pyplot.gca(),lags=40)
pyplot.subplot(212)
plot_pacf(diff2, ax=pyplot.gca(),lags=40)
pyplot.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model_sarimax = SARIMAX(df_train['Daily'] , order=(2,1,2) , seasonal_order=(2,1,2,2))
result_sarimax = model_sarimax.fit()
result_sarimax.summary()
```

```python
forecast = result_sarimax.forecast(steps=41,dynamic=True)
#print(forecast)

#plot the predictions for validation set
plt.plot(df_train['Daily'], label='Train')
plt.plot(df_test['Daily'], label='Test')
plt.plot(forecast, label='Prediction', color='black')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print(np.sqrt(mean_absolute_error(y_true=df_test['Daily'],y_pred=forecast)))
```

```python
from pmdarima import auto_arima
model = auto_arima(df_train['Daily'], start_p=0, start_q=0,max_p=6, max_q=6,max_P=6, max_Q=6,m=10,max_d=6,max_D=6,start_P=0, seasonal=True,d=1, D=1, trace=True,error_action='ignore',suppress_warnings=True,stepwise=True)
model.fit(df_train['Daily'])
forecast_a = model.predict(n_periods=len(df_test['Daily']))
forecast_a = pd.DataFrame(forecast_a,index = df_test['Daily'].index,columns=['Prediction'])
#plot the predictions for validation set
plt.plot(df_train['Daily'], label='Train')
plt.plot(df_test['Daily'], label='Test')
plt.plot(forecast_a, label='Prediction', color='black')
plt.xticks(rotation=90)
plt.show()
```

```python
model.summary()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print(np.sqrt(mean_absolute_error(y_true=df_test['Daily'],y_pred=forecast_a)))
```

```python
p_values=range(0,6)
d_value=range(0,6)
q_values=range(0,6)

import warnings
warnings.simplefilter('ignore')
for p in p_values:
    for d in d_value:
        for q in q_values:
            order=(p,d,q)
            se_od=(p,d,q,2)
            predications=list()
            try:
                model_sarimax=SARIMAX(df_train['Daily'],order=order,seasonal_order=se_od)
                result_sarimax = model_sarimax.fit()
                forecast = result_sarimax.forecast(steps=41,dynamic=True)
                error=np.sqrt(mean_absolute_error(y_true=df_test['Daily'],y_pred=forecast))
                print('SARIMA%s %s RMSE=%.2f'%(order,se_od,error))
            except:
                continue
                    
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model_sa = SARIMAX(df_train['Daily'] , order=(5,1,5) , seasonal_order=(5,1,5,2))
result_sa = model_sa.fit()
```

```python
forecast_sa = result_sa.forecast(steps=41,dynamic=True)
#print(forecast)

#plot the predictions for validation set
plt.plot(df_train['Daily'], label='Train')
plt.plot(df_test['Daily'], label='Test')
plt.plot(forecast_sa, label='Prediction', color='black')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print(np.sqrt(mean_absolute_error(y_true=df_test['Daily'],y_pred=forecast_sa)))
```

```python
result_sa.plot_diagnostics(figsize=(15,12))
plt.show()
```

```python
# Get forecast 100 months ahead in future , this overlaps with our testing dates 
pred_uc = result_sa.get_forecast(steps=61)

# Get confidence intervals of forecasts
pred_ci = pred_uc.conf_int()
```

```python
ax = df_train['Daily'].plot(label='observed', figsize=(20, 10))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Confirmed')

plt.legend()
plt.show()
```
